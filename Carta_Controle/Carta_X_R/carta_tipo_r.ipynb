```python
# Módulos Importantes

import numpy as np
import matplotlib.pyplot as plt
import math as mt
import pandas as pd
```

```python
df = pd.read_csv("medicoes.csv", sep=",", decimal=",")
df
```

```python
# Análizando a Quantidade de Amostras na Tabela

colunas = df.count()
colunas = list(colunas)
print(colunas)

i = 1
for i in range(len(colunas)):
    if colunas[i] ==  colunas[i-1]:
        pass
    else:
        print("ERRO:!! Número de Elementos Diferente")

numero_total_amostras = colunas[1]*len(colunas)

tabela = df
media = df.mean(axis=1)
tabela["Média"] = media 

```

```python
# Aplitude de Cada Linha

max_linha = df.max(axis=1)
min_linha = df.min(axis=1)
amplitude = max_linha - min_linha
tabela["Amplitude"] = amplitude
```

```python
# Valores Importantes

x_barra = media.mean(axis=0)
r_barra = amplitude.mean(axis = 0)
d2 = float(input(print("Entre com o valor da Constante D2")))
d3 = float(input(print("Entre com o valor da Constante D3")))
desv_pad = r_barra/d2

```

```python
# Limites 
limite_superior_x = x_barra + (3*desv_pad)/mt.sqrt(len(colunas))
limite_inferior_x = x_barra - (3*desv_pad)/mt.sqrt(len(colunas))
limite_central_x  = x_barra 
limite_superior_r = x_barra + (3*d3*desv_pad)
limite_inferior_r = x_barra - (3*d3*desv_pad)
limite_central_r  = r_barra 
tabela["LSC_X"] = limite_superior_x
tabela["LCC_X"] = limite_central_x
tabela["LIC_X"] = limite_inferior_x
tabela["LsC_R"] = limite_superior_r
tabela["LCC_R"] = limite_central_r
tabela["LIC_R"] = limite_inferior_r
```

```python
# Carta de Controle do Tipo X
plt.figure(figsize=(25,15))
plt.plot(tabela["LSC_X"],"--")
plt.plot(tabela["LCC_X"], "--")
plt.plot(tabela["LIC_X"], "--")
plt.plot(tabela["Média"],'go-')
plt.grid(True)
plt.ylim(110,125)
plt.xlim(0,25)
plt.title("Carta de Controle Tipo T")
plt.show()
tabela
```

```python
# Carta de Controle do Tipo R
plt.figure(figsize=(25,15))
plt.plot(tabela["LsC_R"],"--")
plt.plot(tabela["LCC_R"], "--")
plt.plot(tabela["LIC_R"], "--")
plt.plot(tabela["Amplitude"],'go-')
plt.grid(True)
plt.ylim(0,50)
plt.xlim(0,25)
plt.title("Carta de Controle Tipo R")
plt.show()
```
